## Q2: What is the success probability? (4 pts)

**For N=4 with 1 iteration:**

## Q4: Why is this faster than classical search? (7 pts)

### âœ… ANSWER:

**Classical Search:**
- Must check items **one by one**
- Worst case: check all N items
- Average: N/2 items
- **Complexity: O(N)**

**Grover's Quantum Search:**
- Uses **superposition** to query all items simultaneously
- **Oracle** marks target by phase flip
- **Amplitude amplification** increases probability of correct answer
- **Complexity: O(âˆšN)**

**Key Quantum Advantages:**

1. **Superposition:** Can be in all N states at once (classical: one at a time)

2. **Amplitude Amplification:** Oracle + diffusion operator:
   - Constructively interfere for correct answer
   - Destructively interfere for wrong answers

3. **Quantum Interference:** Wrong answers cancel out, right answer amplifies

**Example:**  
For N=1,000,000:  
- Classical: ~500,000 queries  
- Grover: ~785 queries  
- **636Ã— faster!**

**Note:** This is **quadratic speedup** (âˆšN), not exponential like Shor's algorithm.

## Q2: Does the original qubit still have the state after teleportation? (5 pts)

### âœ… ANSWER: **NO**

**Reasons:**

**1. No-Cloning Theorem:**  
Quantum mechanics **forbids** copying unknown quantum states. If both had it, we'd have a clone â†’ violates fundamental principle.

**2. Measurement Destroys:**  
Teleportation requires Alice to **measure** original qubit. This measurement **collapses** the state, destroying original superposition.

**3. Information Conservation:**  
Quantum information is **transferred**, not copied. After:
- Original qubit: Collapsed state (no superposition)
- Destination qubit: Has original state (after corrections)

**Analogy:**  
Like Star Trek teleportation - original "disassembled" as copy appears. It's **transfer**, not duplication.

## Q3: Can we teleport faster than light? Why or why not? (6 pts)

### âœ… ANSWER: **NO**

**Quantum teleportation does NOT allow faster-than-light (FTL) communication.**

**Reasons:**

**1. Classical Channel Required:**  
Alice must send **2 classical bits** via normal channel (phone, radio, etc.). This is **limited by speed of light c**.

**2. Bob Can't Use State Without Classical Info:**  
Even though quantum state changes instantly on Bob's side, Bob doesn't know which **corrections** to apply until classical message arrives.

**3. Measurement Outcomes Are Random:**  
Alice's measurements give **random** results. Bob sees only randomness until he knows Alice's results â†’ no information extracted FTL.

**4. No Causality Violation:**  
- Entanglement correlation â‰  information transfer
- Information transfer requires classical communication
- Classical communication â‰¤ c

**Timeline:**
- t=0: Alice measures (instant quantum correlation)
- t=d/c: Classical bits arrive at Bob
- t>d/c: Bob applies corrections and can use state

**Bottom line:**  
Teleportation limited by **classical communication speed**, preserving special relativity!

**Grading:**

- 5 pts: NO + mentions classical communication
